```python
z = np.matrix(rnd.uniform(0, 1, Nt*Nr)).reshape([Nt,Nr])

x = np.sign(p - z)

s = np.zeros([Nt+1, Nr])

for k in range(Nt):
    s[k+1, :] = s[k, :] + x[k, :]

for r in range(Nr):
    plt.plot(s[:,r])
```

# Lognormal price process

```python
def MC_lognormal(Nt, Nr, sigma = 0.3, mu = 0.1, dt = 1/252):
    
    z = np.matrix(rnd.normal(0, 1, Nt*Nr)).reshape([Nt,Nr])
    r = mu*dt + z * sigma *np.sqrt(dt)

    s = np.zeros([Nt+1, Nr])
    for t in range(Nt):
        s[t+1, :] = s[t, :] + r[t, :]

    P = np.exp(s)
    return P
```

```python
Nt = 252
Nr = 5

P = MC_lognormal(Nt, Nr)
    
for run in range(Nr):
    plt.plot(P[:,run])
```

# Simulating AR(1) process

```python
rnd.RandomState().normal(0, 1, [3,2])
```

```python
def MC_AR1(Nt, Nr, lmbd = 0.4, sigma = 0.3, mu = 0.1, dt = 1/252):

    epsilon = rnd.RandomState().normal(0, sigma*np.sqrt(dt), [Nt, Nr])
    R = np.zeros([Nt+1, Nr])
    P = np.ones([Nt+1, Nr])

    for t in range(1, Nt+1):
        R[t,:] = (1+lmbd)*mu*dt - lmbd*R[t-1,:] + epsilon[t-1,:]
        P[t,:] = P[t-1,:] * np.exp(R[t,:])
    
    return R, P
```

```python
Nt = 252
Nr = 1

lmbd = 0.4
sigma = 0.3
mu = 0.1
dt = 1/252

R, P = MC_AR1(Nt, Nr, lmbd = lmbd, sigma = sigma, mu = mu, dt = dt)
```

```python
for run in range(Nr):
    plt.plot(P[:,run])
```

```python
acf = np.zeros([Nt+1, Nr])

x_ovrflw = int(np.ceil(min(10 * np.log10(Nt+1), Nt+1)))
x = range(x_ovrflw)

y = np.power(-lmbd, x)# * np.square(sigma) / (1 - np.square(lmbd))

for run in range(Nr):
    acf[:x_ovrflw,run] = stattools.acf(np.log(1+R[:,run])).transpose()
    plt.figure()
    plt.scatter(x, acf[:x_ovrflw,run])
    plt.scatter(x, y)
    plt.axhline(0)
    plt.show()
```
